```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pprint
```

```python
df = pd.read_csv("dataset.csv").copy()
df.head()
```

```python
df.describe(include="all")
```

<center><h1>Column Variables</h1></center>

```python
def columns_visualozation(column):
    counts = df[column].value_counts()
    percent = df[column].value_counts(normalize=True) * 100
    print(pd.DataFrame({"Count":counts,"Percentage":percent}).head(10))
    
    percentage_nan = df[column].isnull().sum()/(df.shape[0])*100
    print("Missing Values Percentage",percentage_nan) 
    
    values = list(df[column].value_counts().index)
    count = list(df[column].value_counts().values)

    others = sum(count[10:])
    plt.subplots(figsize=(5,5))
    plt.barh(values[:10][::-1],count[:10][::-1],color="orange",alpha=0.7)
    plt.xlabel("COUNT")
    plt.xticks(size=14)
    plt.yticks(size=14)
    plt.ylabel("TOP 10 VALUES (Most Frequent)")
    plt.title("VALUES v/s COUNT")
    plt.grid()
    plt.show()
```

```python
columns = df.columns[2:]
for column in columns:
    print("********************************************************************************************************")
    print("********************************************************************************************************")
    print("VISUALIZING", column)
    print("********************************************************************************************************")
    print("********************************************************************************************************")
    columns_visualozation(column)
    print()
    print()
```
