```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pprint
```

```python
df = pd.read_csv("dataset.csv").copy()
df.head()
```

```python
df.describe(include="all")
```

<center><h1>"AUSTRALIA - AUSTRALIAN DOLLAR/US$"</h1></center>

```python
counts = df["AUSTRALIA - AUSTRALIAN DOLLAR/US$"].value_counts()
percent = df["AUSTRALIA - AUSTRALIAN DOLLAR/US$"].value_counts(normalize=True) * 100
pd.DataFrame({"Count":counts,"Percentage":percent}).head(10)
```

```python
percentage_nan = df["AUSTRALIA - AUSTRALIAN DOLLAR/US$"].isnull().sum()/(df.shape[0])*100
print("Missing Values Percentage",percentage_nan) 
```

```python
values = list(df["AUSTRALIA - AUSTRALIAN DOLLAR/US$"].value_counts().index)
count = list(df["AUSTRALIA - AUSTRALIAN DOLLAR/US$"].value_counts().values)

others = sum(count[10:])
plt.subplots(figsize=(5,5))
plt.barh(values[:10][::-1],count[:10][::-1],color="orange",alpha=0.7)
plt.xlabel("COUNT")
plt.xticks(size=14)
plt.yticks(size=14)
plt.ylabel("TOP 10 VALUES (Most Frequent)")
plt.title("VALUES v/s COUNT")
plt.grid()
plt.show()
```
